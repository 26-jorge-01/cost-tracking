```python
abastecimient_df_exclude_columns = abastecimiento_df.loc[:, abastecimiento_df.columns.isin(abastecimiento_columns)]
abastecimient_df_exclude_columns.head(2)
```

```python
cuentame_df_exclude_columns = cuentame_df.loc[:, cuentame_df.columns.isin(cuentame_columns)]
cuentame_df_exclude_columns.head(2)
```

```python
# --- NORMALIZAR TODO A MAYUSCULAS ---
for df in [abastecimient_df_exclude_columns, cuentame_df_exclude_columns, cuentame_2025_df]:
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.strip().str.upper()

# --- LLAVE COMPUESTA 1er MERGE (UDS + SERVICIO + UBICACION) ---
left_keys = ['Codigo Unidad Servicio UDS', 'SERVICIO 2026', 'Regional UDS', 'Municipio UDS', 'Centro Zonal UDS']
right_keys = ['Código UDS', 'Nombre Servicio', 'Regional UDS', 'Municipio UDS', 'Centro Zonal UDS']

cuentame_dedup = cuentame_df_exclude_columns.drop_duplicates(subset=right_keys, keep='first')

abastecimiento_cuentame = abastecimient_df_exclude_columns.merge(
    cuentame_dedup,
    left_on=left_keys,
    right_on=right_keys,
    how='left',
    suffixes=('', '_cuentame'),
    indicator=True)

for c in [c for c in abastecimiento_cuentame.columns if c.endswith('_cuentame')]:
    abastecimiento_cuentame.drop(columns=[c], inplace=True)

abastecimiento_no_cruza_comp = abastecimiento_cuentame[abastecimiento_cuentame['_merge'] == 'left_only'][abastecimiento_columns].copy()
abastecimiento_cuentame = abastecimiento_cuentame[abastecimiento_cuentame['_merge'] != 'left_only']
abastecimiento_cuentame.drop(columns=['_merge'], inplace=True)

# --- FALLBACK: las no match, solo por UDS contra cuentame ---
cuentame_uds = cuentame_df_exclude_columns.drop_duplicates(subset='Código UDS', keep='first')
fb = abastecimiento_no_cruza_comp.merge(
    cuentame_uds,
    left_on='Codigo Unidad Servicio UDS',
    right_on='Código UDS',
    how='left',
    suffixes=('', '_cuentame'),
    indicator=True)

for c in [c for c in fb.columns if c.endswith('_cuentame')]:
    fb.drop(columns=[c], inplace=True)

abastecimiento_no_cruza_uds = fb[fb['_merge'] == 'left_only'][abastecimiento_columns].copy()
fb = fb[fb['_merge'] != 'left_only']
fb.drop(columns=['_merge'], inplace=True)
abastecimiento_cuentame = pd.concat([abastecimiento_cuentame, fb])
abastecimiento_cuentame.rename(columns=rename_dict, inplace=True)

# --- ULTIMO FALLBACK: cuentame_2025 solo por UDS ---
cue2025_uds = cuentame_2025_df[cuentame_2025_columns].drop_duplicates(subset='CodigoUnidadServicioUDS', keep='first')
abastecimiento_cuentame_2025 = abastecimiento_no_cruza_uds.merge(
    cue2025_uds,
    left_on='Codigo Unidad Servicio UDS',
    right_on='CodigoUnidadServicioUDS',
    how='left',
    suffixes=('', '_cuentame'),
    indicator=True)

for c in [c for c in abastecimiento_cuentame_2025.columns if c.endswith('_cuentame')]:
    abastecimiento_cuentame_2025.drop(columns=[c], inplace=True)
abastecimiento_cuentame_2025.drop(columns=['_merge'], inplace=True)
abastecimiento_cuentame_2025.rename(columns=rename_dict, inplace=True)

abastecimiento_cuentame = pd.concat([abastecimiento_cuentame, abastecimiento_cuentame_2025])
print(f'Registros finales: {len(abastecimiento_cuentame)} (original: {len(abastecimiento_df)})')
len(abastecimiento_cuentame)
```

```python
import win32com.client as win32
import numpy as np

formato_path = base_path + formato

col_map = {
    'Número de Zona': 'Número de Zona',
    'Número de contrato ': 'Número de contrato ',
    'NIT_EntidadContratista': 'Identificación EC',
    'Entidad Contratista\n2026': 'Entidad Contratista\n2026',
    'Servicio 2026': 'Servicio 2026',
    'Regional UDS': 'Regional UDS',
    'Departamento UDS': 'Departamento UDS',
    'Municipio UDS': 'Municipio UDS',
    'Centro Zonal UDS': 'Centro Zonal UDS',
    'Codigo Unidad Servicio UDS': 'Codigo Unidad Servicio UDS',
    'Unidad Servicio UDS': 'Unidad Servicio UDS',
    'Barrio UDS': 'Barrio UDS',
    'Direccion UDS': 'Direccion UDS',
    'Estado UDS': 'Estado UDS',
    'Vigencia UDS': 'Vigencia UDS',
    'Cupos UDS ofertados 2026': 'Cupos UDS ofertados 2026',
}

col_const = {
    'Modalidad\n2026': 'INTEGRAL',
}

def _col_letter(n):
    letters = ''
    while n > 0:
        n -= 1
        letters = chr(65 + n % 26) + letters
        n //= 26
    return letters

# Mapear nombre de columna del formato a su letra de Excel
formato_header = pd.read_excel(formato_path, sheet_name='zonificacion', nrows=0).columns.tolist()
col_name_to_letter = {name: _col_letter(i+1) for i, name in enumerate(formato_header)}

# --- Escribir con xlwings ---
import xlwings as xw
import os

os.system('attrib -R "{}"'.format(formato_path))

app = xw.App(visible=False)
wb = app.books.open(formato_path)
ws = wb.sheets['zonificacion']

nrows = len(abastecimiento_cuentame)

for src_col, tgt_col in col_map.items():
    col_letter = col_name_to_letter[tgt_col]
    ws.range(f'{col_letter}2:{col_letter}{nrows+1}').value = None
    values = abastecimiento_cuentame[src_col].tolist()
    values = [[None if pd.isna(v) else v] for v in values]
    ws.range(f'{col_letter}2').value = values
    print(f'  Escrita columna {tgt_col} ({col_letter})')

for tgt_col, const_val in col_const.items():
    col_letter = col_name_to_letter[tgt_col]
    ws.range(f'{col_letter}2:{col_letter}{nrows+1}').value = None
    ws.range(f'{col_letter}2').options(transpose=True).value = [const_val] * nrows
    print(f'  Escrita columna constante {tgt_col} ({col_letter})')

wb.save()
wb.close()
app.quit()

print(f'Formato actualizado correctamente: {nrows} filas escritas en {formato_path}')
```
