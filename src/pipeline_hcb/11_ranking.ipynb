```python
import pandas as pd
```

```python
base_path = '../../data/insumos 19 junio/'
source_name = '20260615_Costeo_Adición_HCB_Julio_Octubre_costeo_actualizadoV5 1.xlsx'
```

```python
contratos = pd.read_excel(base_path + source_name, sheet_name='COSTEO', dtype={'No. Contrato': str})
```

```python
contratos.head(2)
```

```python
contratos['Regional UDS'] = contratos['Regional UDS'].str.title()
len(contratos['Regional UDS'].value_counts())
```

```python
sorted(contratos.columns)
```

```python
target_columns = ['Regional UDS', 'No. Contrato']
contratos_res = contratos[target_columns].drop_duplicates().reset_index(drop=True)

print('contratos: ', len(contratos))
print('contratos_res: ', len(contratos_res))
```

```python
contratos_res['PACCO'] = ''
contratos_res['CDP'] = ''
contratos_res['SITCO'] = ''
contratos_res['Estudios Previos'] = ''
contratos_res['Documentos del Operador'] = ''
contratos_res['Radicación Jurídica'] = ''
contratos_res['SECOP'] = ''
contratos_res['Póliza'] = ''
contratos_res['RP'] = ''
contratos_res['Referencia de Contrato'] = ''
contratos_res['Categoria EAS'] = ''

contratos_res.head(2)
```

```python
for regional in contratos_res['Regional UDS'].unique():
    contratos_res[contratos_res['Regional UDS'] == regional].to_excel(base_path + 'Ranking/' + f'ranking_{regional}.xlsx', index=False)
```
