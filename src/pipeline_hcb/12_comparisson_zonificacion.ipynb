```python
import pandas as pd
```

```python
base_path = '../../data/insumos 19 junio/'
original_source_name = '20260615_Costeo_Adición_HCB_Julio_Octubre_costeo_actualizadoV5 1.xlsx'
new_source_name = '20260619_Costeo_Adición_HCB_Julio_Octubre_costeo_actualizadoV10 1.xlsx'
```

```python
original_source = pd.read_excel(base_path + original_source_name, sheet_name='COSTEO', dtype={'No. Contrato': str})
print(len(original_source))
original_source.head(2)
```

```python
new_source = pd.read_excel(base_path + new_source_name, sheet_name='COSTEO', dtype={'No. Contrato': str})
print(len(new_source))
new_source.head(2)
```

```python
original_source['Regional UDS'] = original_source['Regional UDS'].str.title()
new_source['Regional UDS'] = new_source['Regional UDS'].str.title()

original_source_contracts_by_regional = original_source[['No. Contrato', 'Regional UDS']].drop_duplicates()
new_source_contracts_by_regional = new_source[['No. Contrato', 'Regional UDS']].drop_duplicates()

print(len(original_source_contracts_by_regional))
print(len(new_source_contracts_by_regional))
```

```python
original_new_contracts = pd.merge(
    left=original_source_contracts_by_regional, right=new_source_contracts_by_regional,
    how='outer', on=['No. Contrato', 'Regional UDS'], indicator=True)

original_new_contracts['_merge'].value_counts()
```

```python
original_new_contracts[original_new_contracts['_merge'] == 'left_only'].reset_index()[['No. Contrato', 'Regional UDS']]
```

```python
original_new_contracts[original_new_contracts['_merge'] == 'right_only'].reset_index()[['No. Contrato', 'Regional UDS']]
```
